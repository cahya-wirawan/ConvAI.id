```python
from tqdm import tqdm
from transformers import MarianMTModel, MarianTokenizer
import torch
```

```python
class Translation():
    def __init__(self, model_name):
        self.device = "cuda" if torch.cuda.is_available() else "cpu"
        self.tokenizer = MarianTokenizer.from_pretrained(model_name)
        self.model = MarianMTModel.from_pretrained(model_name).to(self.device)

    def translate(self, source_text):
        inputs = self.tokenizer(source_text, return_tensors="pt", padding=True).to(self.device)
        translated = self.model.generate(**inputs)
        translation = [self.tokenizer.decode(t, skip_special_tokens=True) for t in translated]
        return translation
```

```python
model_name = "Wikidepia/marian-nmt-enid"
translator = Translation(model_name)
```

```python
sentences = [
             "I am hungry, but I don't have money to buy food",
             "My friend goes to school every morning with bus"
]
```

```python
%%time

translator.translate(sentences)
```
